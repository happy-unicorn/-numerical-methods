```python
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from pylab import rcParams
from IPython.display import clear_output
```

```python
rcParams['figure.figsize'] = (15, 15)
```

### Код игры

```python
def generate_field(n):
    return np.random.choice([0, 1], n * n).reshape(n, n)
    
    
def plot_result(field):
    clear_output(wait=True)
    sns.heatmap(field, square=True, xticklabels=False, yticklabels=False, cbar=False, vmin=0, vmax=1)
    plt.show()
    
    
def update(field, n):
    tmp = field.copy()
    
    for i in range(n):
        for j in range(n):
            total = int(
                field[i, (j - 1) % n] + 
                field[i, (j + 1) % n] +       
                field[(i - 1) % n, j] + 
                field[(i + 1) % n, j] +        
                field[(i - 1) % n, (j - 1) % n] + 
                field[(i - 1) % n, (j + 1) % n] +        
                field[(i + 1) % n, (j - 1) % n] + 
                field[(i + 1) % n, (j + 1) % n]
            )
 
            if field[i, j]  == 1:
                if (total < 2) or (total > 3):
                    tmp[i, j] = 0
            else:
                if total == 3:
                    tmp[i, j] = 1
    
    field[:] = tmp[:]
    

def run_game(n=100, iterations=100):
    field = generate_field(n)
    for i in range(iterations):
        update(field, n)
        plot_result(field)
```

```python
run_game(iterations=5)
```
